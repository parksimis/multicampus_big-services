### 1) Line Plot

* `plot()`
    * [documentation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.plot.html)
    * 선을 그리는 함수
    * 데이터가 시간, 순서 등에 따라 변화를 보여주기 위해 사용
    * x축 값 생략 시 자동 생성(0부터 시작하는 len(y)까지 정수)
    
#### - 관련 method
- `figure()` : 그래프 크기 설정
- `title()` : 제목 출력
- `xlim()` : x축 범위
- `ylim()` : y축 범위
- `xticks()`/ `yticks()` : 축과 값을 잇는 실선 
- `legend()` : 범례
- `xlabel()` / `ylabel()` : x / y축 라벨(값)
- `grid()` : 그래프 배경으로 격자무늬 사용


#### - 자주 사용되는 스타일 속성(약자로도 표기 가능)
- `color(c)` : 선 색
- `linewidth(lw)` : 선 굵기
- `linestyle(ls)` : 선 스타일
- `marker` : 마커 종류
- `markersize(ms)` : 마커 크기
- `markeredgecolor(mec)` : 마커 선 색
- `markeredgewidht(mew)` : 마커 선 굵기
- `markerfacecolor(mfc)` : 마커 내부 색

```python
# 기본적인 lineplot 그리기 - x축은 자동 지정
plt.plot([1, 4, 9, 16])
plt.show()
```

#### `subplot()` 
* [documnetation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.subplot.html)
* 하나의 윈도우(figure) 안에 여러 개의 플롯을 배열 형태로 표시, 그리드 형태의 Axes 객체들을 생성한다.
    * 형식 : subplot(인수1, 인수2, 인수3)
        * 인수 1과 인수2는 전체 그리드 행렬 모양 지시
        * 인수 3 : 그래프 위치 번호
    * Ex. 2행 2열의 첫번째 위치
        * `subplot(2, 2, 1)`가 원칙이나 줄여서
        * `subplot(221)`로 가능

```python
np.random.seed(1)

# 2 * 2 형태의 네 개의 plot

plt.subplot(221)
plt.plot(np.random.randn(5))
plt.title('axes1')

plt.subplot(222)
plt.plot(np.random.randn(5))
plt.title('axes2')

plt.subplot(223)
plt.plot(np.random.randn(5))
plt.title('axes3')

plt.subplot(224)
plt.plot(np.random.randn(5))
plt.title('axes4')
plt.tight_layout()
plt.show()
```

#### `subplots()`
* [documentation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.subplots.html)
* 여러 개의 Axes 객체를 동시 생성해주는 함수
* 2차원 ndarray 형태로 Axes 객체를 반환함 (행렬 형태로 반환)
* 두 개의 반환값이 있고, 첫번째 반환값은 그래프 전체 객체의 이름(사용 X), 두 번째 반환값에 Axes 객체 반환

```python

fig, axes = plt.subplots(2,2)

np.random.seed(1)

axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title('axes 1')

axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title('axes 2')

axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title('axes 3')

axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title('axes 4')

plt.tight_layout()
plt.show()
```

```python
# 범례 표시하기
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.grid()
plt.legend(loc=2)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Example of sin/cosine wave')
plt.show()
```

### 2) bar chart 
* 막대 그래프 그리기

#### `bar()`
* [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)


```python
y = [2, 3, 1, 4]
x = np.arange(len(y))

z = [2, 3]
s = [0, 1]

e = [1, 4]
h = [2, 3]

xlabel = ['가', '나', '다', '라']

plt.title('Bar chart')
plt.bar(x, y)
plt.plot(s, z, 'r--o') # s, z 위치에 선을 그음
plt.plot(h, e, 'g--o')
plt.yticks(sorted(y))
plt.xticks(x, xlabel)
plt.xlabel('가나다라')
plt.ylabel('빈도 수')
# text()로 글자 찍을 수 있음
plt.text(-0.1, 1, r'test')

plt.show()
```

####  `barh()`
* 가로 막대 그래프
* [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.barh.html)

```python
# 연습용 데이터 생성

np.random.seed(1)
people = ['몽룡', '춘향', '방자', '향단']
y = np.arange(len(people))
y
# rand() 0부터 1 사이의 유니폼 분포에서 난수 생성
performance = 3 + 10*np.random.rand(len(people))
performance
```

```python
# 가로 막대그래프
plt.title('Barh Chart')
plt.barh(y, performance, alpha=.4)
plt.yticks(y, people)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.grid()
plt.show()
```

### 3) scatter plot
#### `scatter()`
* [documentation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.scatter.html)
* 분산형 그래프

```python
t = np.arange(10)
y = np.random.randint(1, 10, size=(10,))

# marker 옵션으로 marker 지정 가능
plt.scatter(t, y, marker='^')
plt.show()
```

* 버블 차트 : 점 하나의 크기 또는 색상을 이용해서 서로 다른 데이터 값을 표시하는 그래프
    * parameter
        * `s` : size
        * `c` : color

```python
np.random.seed(1)
N = 30
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi*(15*np.random.rand(N))**2
x;y1;y2;y3


plt.title('Bubble Chart')
plt.scatter(x, y1, c=y2, s=y3)
plt.colorbar()
plt.show()
```

```python
# colorbar 표현하기 -  color
# t값에 따라서 색이 다르게 표시되고 있음.
# 진할수록 작은 값, 옅을수록 큰 값

plt.scatter(t, y, s=50, c=t, marker='>')
plt.colorbar()
plt.show()
```

### 5) `boxplot()`
* 박스 플롯
* [documentation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.boxplot.html)
  
 

```python
# 연습용 데이터 생성
#데이터 3개 생성
s1 = np.random.normal(loc=0,scale=1,size=1000)
s2 = np.random.normal(loc=5,scale=0.5,size=1000)
s3 = np.random.normal(loc=10,scale=2,size=1000)


# line 그래프를 이용해서 데이터 차이 확인

plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
# 박스 그래프

plt.boxplot((s1, s2, s3))
plt.grid()
plt.show()
```

### 6) `pie()`
* `pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=0, 0, frame=False, rotatelabels=False, *, normalize=None, data=None)`
* [documentation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.pie.html?highlight=pie#matplotlib.pyplot.pie)
* 카테고리 별 값의 상대적인 비교를 할 때 주로 사용하는 차트
* 원의 형태를 유지할 수 있도록 다음 명령(`plt.axis('equal')`)을 실행해야 한다.
* 콘솔에서는 별 다른 변화가 없지만, plot 창에서는 필요하다.

```python
#예제 데이터 생성

labels=['개구리','돼지','개','통나무']
size=[15,30,45,10]
colors=['yellowgreen','gold','lightskyblue','lightcoral']

# explode를 통해 조각을 띄어서 보여줄 수 있음
explode=(0, 0.4, 0, 0.5)
```

```python
plt.title('Pie Chart')
plt.pie(size, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=180)
plt.axis('equal')
plt.show()
```
