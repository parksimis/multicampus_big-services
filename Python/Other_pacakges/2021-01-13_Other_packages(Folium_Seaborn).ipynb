### 1. 초기 지도 객체 생성
* `.map()`을 사용해 중심좌푯값을 지정함으로써 간단하게 생성이 가능하다.
* 위도와 경도를 data로 지도를 그려줌
    * 크롬에서는 출력까지 가능(Explore에서는 저장만 가능)
    
    
* `Map()`
    * [documenatation](https://python-visualization.github.io/folium/modules.html?highlight=map#folium.folium.Map)
    * `location` : 위경도
    * `zoom_start` : 초기 확대 비율 정의
    * `tiles` : 지도 스타일 변경 (Stame Tone / Stame Terrain)

```python
map_osn = folium.Map(location=[37.3004794,127.0357158], zoom_start=13)
map_osn
```

```python
map_osn = folium.Map(location=[37.3004794,127.0357158], zoom_start=13, tiles='Stamen Toner')
map_osn
```

### 2. Marker 설정(popup 설정도 가능)
* Marker(특정 위치를 표시하는 표식)
* popup(마커 클릭 시 나타나는 정보) : 환경에 따라 지원 안되는 경우도 있음.
  
  
#### 1) 마커 생성
* `folium.Marker()` : 
    * [documentation](https://python-visualization.github.io/folium/modules.html?folium.map.Marker)
    * parameter
        * `location` :위, 경도 값 리스트
        * `popup`: popup 문자
        * `icon` : 마커의 모양
    
#### 2) 생성 마커 부착
* `folium.Marker().add_to(지도객체)`

```python
map_osn = folium.Map(location=[37.3004794,127.0357158], zoom_start=13)
folium.Marker(location=[37.3004794,127.0357158], popup='Kyonggi University', icon=folium.Icon(icon='cloud', color='red')).add_to(map_osn)
map_osn
```

#### 연습

* 특정 위치의 위경도를 파악해서 지도를 표시하고 마커 생성해 찍어보기

```python
# 광화문 세종대왕 동상 지도 표시
map_seoul = folium.Map(location=[37.5762187,126.9748017], zoom_start=15)
map_seoul
```

### `seaborn.despine()`
* 축선 표시 여부를 결정하는 함수
* [documentation](https://seaborn.pydata.org/generated/seaborn.despine.html)
* parameter
    * `right` = T/F
    * `left` = T/F
    * `top` = T/F
    * `bottom` = T/F
    * 각 속성값이 True : 표시하지 않음 / False면 표시함
    * `offset` = x, y 축의 간격을 설정

```python
#sns.set_style('darkgrid')

plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine(right=False, left=True, top=False, bottom=True)
plt.show()
```

#### seaborn의 기본 dataset : tips 활용
* 요일별 점심 저녁 식사 금액과 팁을 정리한 Dataset
* 성별, 흡연 여부가 구별되어 있음.
* `sns.load_dataset('dataset_name')`

```python
tips = sns.load_dataset('tips')
tips.head()
tips.tail()
tips.shape
```

##### 1) tips dataset을 이용한 boxplot 그리기
* `sns.boxplot()`
    * [documentation](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

```python
sns.set_style('whitegrid')
sns.boxplot(x=tips['total_bill'])
```

위의 그래프는 모든 요일의 전체 영수증 금액의 분포를 본 것이다. 각 요일별 영수증 금액의 분포를 볼 수도 있다.  
적게는 10달러 아래, 많게는 40달러 이상의 이상치들도 있고, 대부분 20 달러를 기준으로 좌우로 분포하고 있는 것을 알 수 있다. 
  
##### 2) 요일별 영수증 금액의 분포 확인
* 색상 변경을 위해 `palette` 인수를 사용 가능


```python
sns.boxplot(x=tips['day'], y=tips['total_bill'], palette='Set3')
sns.despine(offset=10)
plt.show()
```

대부분 비슷한 분포를 보이지만, 토요일은 다른 날짜에 비해 간격이 넓게 분포하고 있는 것을 알 수 있다.  
  
* `hue` : 카테고리 값을 가지는 변수의 이름을 지정하면, 카테고리에 따라서 그룹을 구성해서 그래프에 반영한다.

흡연 여부에 따라서 영수증 금액의 분포를 확인해보고자 하면 `hue=tips['smoker']`를 사용  
또는, data에 사용할 data의 변수명을 지정해주면 `hue='smoker`로 변수명만 적어줄 수 있음(다른 인수들에도 동일 적용)

```python
#sns.boxplot(x=tips['day'], y=tips['total_bill'], palette='Set3', hue=tips['smoker'])
sns.boxplot(x='day', y='total_bill', palette='Set3', hue='smoker', data=tips)
# 위 아래 동일한 결과가 나옴
sns.despine(offset=10)
plt.show()
```

토요일은 흡연/비흡연자가 전체적으로 넓은 분포를 가지는 것을 알 수 있다.  
금요일은 흡연자가 비흡연자보다 더 넓은 분포를 가진다.  
이외의 요일에는 비슷한 분포를 가지는 것으로 보인다.  
  
##### 연습 : 요일별 팁의 분포를 비교하되, 성별에 따라 분포를 나누어 표시해보자.

### `seaborn.swarmplot()`
* 데이터가 모여있는 정도를 보여주는 plot
* 위의 boxplot과 비슷한 인수를 가진다.
* [documentation](https://seaborn.pydata.org/generated/seaborn.swarmplot.html)

```python
sns.swarmplot(data=tips, x='day', y='total_bill', palette='Set3')
sns.despine()
plt.show()
```

`swarmplot()`은 위의 `boxplot()`을 scatter 형태로 보여주는 것이라고 이해하면 쉽다.  
비슷한 형태를 띄기 때문에, 두 그래프를 한 그래프에 보여주는 경우가 많다.

```python
sns.boxplot(x='day', y='total_bill', palette='Set3', data=tips)
sns.swarmplot(data=tips, x='day', y='total_bill', palette='Set1')
sns.despine()
plt.show()
```

두 데이터를 한꺼번에 보여줘, 분포와 실제 데이터의 위치를 동시에 보여주는 것이 가능하다.

### `seaborn.lmplot()`
* scatter 그래프를 그릴때 회귀선을 기본으로 출력해주는 그래프 함수
* [documentation](https://seaborn.pydata.org/generated/seaborn.lmplot.html)
* parameter
    * `height` : 그래프 크기
    * `fit_reg` : False이면, 회귀선 생략하게됨(default=True)
    * `hue` : 그룹을 나누어 보여줌
    * `col` / `row` : 나누어진 그룹을 column(옆) / row(아래)에 보여줌

```python
sns.lmplot(data=tips, x='total_bill', y='tip', fit_reg=True, height=5)
plt.show()
```

total_bill과 tip에 대한 scatter 그래프를 그려주고, 이를 가장 잘 설명해줄 수 있는 회귀선을 그려준다.

* `hue` 인수를 활용하면 데이터를 그룹화하여 그래프를 보여줄 수 있다.

```python
sns.set_style('darkgrid')
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker', height=5)
plt.show()
```

위의 그래프를 보면 흡연 여부에 따라서 다른 그래프와 회귀식을 보여주는 것을 알 수 있다.  
  
##### 연습 
* 영수증 금액과 팁의 관계성을 `lmplot()`으로 표현하되, 성별을 구분해서 별도의 그래프로 표시
* 단 회귀선 표시 X

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='sex', fit_reg=False, height=7)
sns.despine()
plt.show()
```

### `seaborn.heatmap()`

* [documentation](https://seaborn.pydata.org/generated/seaborn.heatmap.html)
* 각 값의 크기를 색의 농도를 통해 표현
* parameter
    * `data` : 표현할 데이터 
    * `annot` : T/F, True면 박스 안에 값을 표현
    * `fmt` : annot 시 포매팅 지정



```python
# 예제 데이터 생성
np.random.seed(1)
uniform_data = np.random.rand(10, 12)
uniform_data.shape

# heatmap 그리기
sns.heatmap(uniform_data, vmin=0, vmax=1)
plt.show()
```

* seaborn의 flights dataset을 이용해 heatmap 그려보기

```python
flights = sns.load_dataset('flights')
flights.shape
flights.head()
flights.tail()
```

```python
# pivot 테이블을 이용해 연도/월별 승객 수로 구조를 변환
# pivot(인덱스, 컬럼명, 계산필드)
# 이를 통해 각 컬럼은 각 연도, index는 월인 승객 수 테이블을 얻을 수 있다.

flight_pivot = flights.pivot(index='month', columns='year', values='passengers')
flight_pivot
```

```python
# seaborn 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
sns.heatmap(flight_pivot, annot=True, fmt='d')
plt.title('월 / 연도별 승객 수')
plt.show()
```
