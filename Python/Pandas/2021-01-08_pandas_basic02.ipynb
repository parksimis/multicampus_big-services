#### 데이터 프레임 만들기
* dict 데이터프레임 만들기
* 열방향 인덱스 `columns=`
* 행방향 인덱스 `index=`

```python
# 필요 라이브러리 import
import pandas as pd
import numpy as np
```

```python
data = {
    '2015' : [9904312, 3448737, 2890451, 2466052],
    '2010' : [9631482, 3393191, 2632035, 2000002],
    '2005' : [9762546, 3512547, 2517680, 2456016],
    '2000' : [9853972, 3655437, 2466338, 2473990],
    '지역' : ['수도권', '경상권', '수도권', '경상권'],
    '2010-2015 증가율' : [0.0283, 0.0163, 0.0982, 0.0141]
}

index = ['서울', '부산', '인천', '대구']
columns = ['지역', '2015', '2010', '2005', '2000', '2010-2015 증가율']

# 데이터프레임 만들기

df = pd.DataFrame(data, index=index, columns=columns)
df
```

* DataFrame의 구성요소 : `values`, `index`, `columns` (객체의 속성으로 구성)

```python
# values - array로 반환
print(df.values)

# columns - index로 반환
print(df.columns)

# index
print(df.index)
```

* DataFrame의 개요 출력 - `info()`
* DataFrame의 통계적 개요 출력 - `describe()`

```python
print(df.info())

display(df.describe())
```

* DataFrame도 이전의 Series에서 마찬가지로 각 방향 인덱스의 name을 정의할 수 있다.

```python
# 행 방향 인덱스 name 정의
df.index.name = '도시'

# 열 방향 인덱스 name 정의
df.columns.name = '특성'

display(df)
```

## 데이터 프레임 생성 예제 2 : df3 생성

* 데이터 : numpy package의 `randn`을 이용해서 난수 n개 생성
* 행방향 인덱스(index) : 날짜 데이터
    * pandas `date_range(시작날짜, periods=생성개수, freq=생성간격)`
    * frequency : W (일주일 단위로 생성)
* 열방향 인덱스(columns) : ['A', 'B', 'C', 'D']


```python
# 1) 행방향 인덱스 생성 : 날짜데이터

dates = pd.date_range('20190101', periods=6, freq='W')
print(dates)

# 2) DataFrame 생성

df3 = pd.DataFrame(np.random.randn(6, 4), index=dates, columns=['A', 'B', 'C', 'D'])
display(df3)
```

```python
# 예제 데이터 파일 생성

data = [[22, 60.1, 170.5, '남', '서울'], 
        [45, 51.3, 157.3, '여', '부산'], 
        [22, 68.3, 180.1, '남', '대구'],
        [33, 88.3, 190.2, '남', '제주'], 
        [27, 48.3, 160.1, '여', '강릉']]
df4 = pd.DataFrame(data, index=['홍길동', '이몽룡', '성춘향','변학도','김연아'], 
                   columns=["나이", "몸무게", "키", "성별" ,"주소"])
df4
```

* pandas DataFrame은 전치를 포함해서 Numpy 2차원 배열의 대부분 속성이나 메소드를 지원한다.
* 전치(Transpose)는 행과 열을 바꾸는 기능
    * 사용방법 : `객체명.T`
    * 원본 데이터를 변경하지 않는다.

```python
# Transpose
display(df4.T)

# 원본 객체가 바뀌지 않는 것을 알 수 있다.
display(df4)
```

```python
# 위의 df을 Transpose하고 df_1 변수에 저장하시오.
# 원본과 비교 출력해보기

display(df)

df1 = df.T
display(df1)
```

### 해당 열이 있으면 내용 갱신, 열이 없으면 추가
* 열 추가 및 갱신 : `df[열이름(key)] = values`
* 열 삭제 - `del` , `drop()` 등이 있음.

```python
# 증가율 변경
df['2010-2015 증가율'] = df['2010-2015 증가율'] * 100
display(df)
```

```python
# 열 추가 - 2005년 대비 2015년 인구 증가율

df['2005-2015 증가율'] = ((df['2015'] - df['2005']) / df['2005'] * 100).round(2)
display(df)
```

```python
# 열 삭제
del df['2010-2015 증가율']
display(df)
```

### DataFrame indexing_basic

* 1. 열 인덱싱
* 2. 인덱서를 사용하지 않는 행 인덱싱
* 3. `[ ]`를 이용해서 인덱싱 할 때 주의할 점 : `[ ]` 기호는 열 위주 인덱싱이 원칙

#### 1. 열 인덱싱
* 열 라벨(컬럼명)을 Key 값으로 생각하고 인덱싱함.
    * 인덱스로 라벨 값을 하나 넣으면 시리즈 객체가 반환 (차원 축소)
    * 라벨의 배열이나 리스트를 넣으면 부분적 df가 반환

```python
# 인덱스로 라벨(열이름)값 1개 사용
print(df['지역'])
print(type(df['지역']))
# 시리즈 객체로 반환하는 것을 알 수 있다.

# 열 1개를 접근하는 경우에는 .연산자 사용 가능 : 객체명.컬럼명
print(df.지역)
```

```python
# 열 1개를 추출할 때, df 형태로 반환받고 싶을 때 (차원 축소 방지)
display(df[['지역']])

# 차원이 축소되지 않고, dataframe 객체로 그대로 반환되는 것을 알 수 있다.
print(type(df[['지역']]))
```

```python
# 열의 부분 추출 - 리스트 형태로 인덱싱 해야함

df[['2010', '2015']]
```

#### 연습문제
* 다음 DataFrame에서 지정하는 데이터를 뽑아내거나, 처리하라

1. 모든 학생의 수학 점수를 시리즈로 나타낸다.
2. 모든 학생의 국어와 영어 점수를 데이터프레임으로 나타낸다.
3. 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.
4. 방자의 모든 영어 점수를 80점으로 수정하고 평균점수도 다시 계산한다.
5. 춘향의 모든 점수를 데이터프레임으로 나타낸다.
6. 향단의 점수를 시리즈로 나타낸다.

```python
data =  {
    '국어' : [80, 90, 70, 30],
    '영어' : [90, 70, 60, 40],
    '수학' : [90, 60, 80, 70]
}

columns = ['국어', '영어', '수학']
index = ['춘향', '몽룡', '향단', '방자']

prac_df = pd.DataFrame(data, index=index, columns=columns)
display(prac_df)
```

```python
# 1) 모든 학생의 수학 점수를 시리즈로 나타내라
display(prac_df['수학'])
```

```python
# 2) 모든 학생의 국어와 영어 점수를 데이터프레임으로 나타내라
display(prac_df[['국어', '영어']])
```

```python
# 3) 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.
prac_df['평균'] = prac_df.mean(axis=1).round(2)
display(prac_df)
```

```python
# 4) 방자의 모든 영어 점수를 80점으로 수정하고 평균점수도 다시 계산한다.
prac_df['영어']['방자'] = 80
prac_df['평균'] = prac_df.mean(axis=1).round(2)

# 원본 변경으로 인한 Warning 메시지 출력 -> 일단은 그대로 둠.
display(prac_df)
```

```python
# 5) 춘향의 모든 점수를 데이터프레임으로 나타낸다.
display(prac_df[:1])
```

```python
# 6) 향단의 점수를 시리즈로 나타낸다.
prac_df.loc['향단']
```
